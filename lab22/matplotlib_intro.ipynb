# Matplotlib Intro

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3])
y = x**2

print(x)
print(y)

plt.scatter(x,y)
plt.show()
```

## Create figure and axes with Pyplot Interface:

```python
fig = plt.figure(facecolor='#900')

fig.add_subplot(221)
plt.scatter(x,y)

fig.add_subplot(222)
plt.plot(x,y)

fig.add_subplot(223)
plt.bar(x,y)

fig.add_subplot(224)
plt.hist(x,y)
```

## Create figure and axes with OOP Interface:

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
fig.set_facecolor('#900')
fig.set_size_inches(10, 5)

ax1 = axes[0][0]
ax2 = axes[0][1]
ax3 = axes[1][0]
ax4 = axes[1][1]

ax1.scatter(x,y)
ax2.plot(x,y)
ax3.bar(x,y)
ax4.hist(x,y)

ax1.set_title('Scatter')
ax3.set_title('Bar')

ax1.set_xlabel('X',loc='right' ,color='yellow', fontsize='14')
ax1.xaxis.set_label_coords(1.05,-0.02)
# ax1.yaxis.set_label_coords(1.05,0.02)
ax1.set_ylabel('Y',loc='top', rotation=0, color='yellow')

plt.tight_layout()


# ticks
x_tick_labels = ax1.get_xticklabels()
print(x_tick_labels)

x_tick_labels[5].set_color('yellow')
```

## Color Values

```python
x = np.arange(1,11)
y1 = x**2
y2 = x**3


fig,[ax1,ax2] = plt.subplots(nrows=2, ncols=1)
# fig.set_facecolor('#RedGreeenBlue')
fig.set_facecolor('#990099')

ax1.set_xlabel('X', fontdict={
		'fontname': 'monospace', # try 'cursive' or 'serif'
		'fontsize': 30,
		'fontweight': 'bold',
		'color': 'red',  
		'backgroundcolor':'#999',
		'verticalalignment': 'bottom',		
	 })
ax1.set_ylabel('Y')

ax1.plot(x,y1,'Dr:',label='y1')
ax1.plot(x,y2,label='y2')
ax1.legend(loc = 'upper left', fontsize = 'large', shadow=True,framealpha = 1)

ax2.scatter(x,y1, marker='D')
ax2.plot(x,y2)

plt.show()
```
