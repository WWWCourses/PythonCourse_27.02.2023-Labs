```python
### Plot 3d line:

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define data
x = [1, 2, 3]
y = [4, 5, 6]
z = [7, 8, 9]

# Plot the line
ax.plot(x, y, z)

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Basic 3D Plot')

# Show the plot
plt.show()
```

```python
### Plot scatter 

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Generate random data
np.random.seed(0)
n = 100
x = np.random.rand(n)
y = np.random.rand(n)
z = np.random.rand(n)

# Plot the scatter plot
ax.scatter(x, y, z, c='r', marker='o')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Scatter Plot in 3D')

# Show the plot
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define the Rosenbrock function
def rosenbrock(x, y):
    return (1 - x)**2 + 100 * (y - x**2)**2

# Generate data for the surface plot
x = np.linspace(-2, 2, 100)
y = np.linspace(-1, 3, 100)

X, Y = np.meshgrid(x, y)
Z = rosenbrock(X, Y)

print(X)
print(Y)
print(Z)

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Surface Plot')

# Show the plot
plt.show()
```
